```python
#Convert to proper data type
df['Year_of_Release']=df['Year_of_Release'].astype('Int64')
df['Critic_Score']=df['Critic_Score'].astype('Int64')
df['Critic_Count']=df['Critic_Count'].astype('Int64')
df['User_Count']=df['User_Count'].astype('Int64')
df['User_Score'] = pd.to_numeric(df['User_Score'], errors= 'coerce')
```

```python
#Check if values are of the float datatype now
df['User_Score'].unique()
```

```python
df.info()
```

```python
df=df.sort_values('Year_of_Release')
```

```python
df
```

```python
sales_per_year = df.groupby('Year_of_Release')[['NA_Sales', 'EU_Sales', 'JP_Sales']].sum().reset_index()
sales_per_year.astype('int64')
```

```python
sales_per_year.columns
```

```python
plt.figure(figsize=(10,6))
plt.plot(sales_per_year['Year_of_Release'], sales_per_year['NA_Sales'], label = 'NA_Sales')
plt.plot(sales_per_year['Year_of_Release'], sales_per_year['EU_Sales'], label = 'EU_Sales')
plt.plot(sales_per_year['Year_of_Release'], sales_per_year['JP_Sales'], label = 'JP_Sales')
plt.legend()
plt.show()
```
