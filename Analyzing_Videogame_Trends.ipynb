```python
#Convert to proper data type
df['Year_of_Release']=df['Year_of_Release'].astype('Int64')
df['Critic_Score']=df['Critic_Score'].astype('Int64')
df['Critic_Count']=df['Critic_Count'].astype('Int64')
df['User_Count']=df['User_Count'].astype('Int64')
df['User_Score'] = pd.to_numeric(df['User_Score'], errors= 'coerce')
```

```python
#Check if values are of the float datatype now
df['User_Score'].unique()
```

```python
df=df.sort_values('Year_of_Release')
```

```python
sales_per_year = df.groupby('Year_of_Release')[['NA_Sales', 'EU_Sales', 'JP_Sales']].sum().reset_index()
sales_per_year.astype('int64')
```

```python
sales_per_year.columns
```

```python
plt.figure(figsize=(10,6))
plt.plot(sales_per_year['Year_of_Release'], sales_per_year['NA_Sales'], label = 'NA_Sales')
plt.plot(sales_per_year['Year_of_Release'], sales_per_year['EU_Sales'], label = 'EU_Sales')
plt.plot(sales_per_year['Year_of_Release'], sales_per_year['JP_Sales'], label = 'JP_Sales')
plt.title('Sales per Region')
plt.legend()
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize= (10, 6))
df.groupby('Genre')['JP_Sales'].sum().sort_values(ascending=False).plot(kind='bar', ax = ax1)
ax1.yaxis.set_major_formatter(plt.FuncFormatter(lambda y, pos: f'${int(y)}M'))
ax1.set_title('Videogame Sales in Japan by Genre')
ax1.set_xticklabels(ax1.get_xticklabels(), rotation=45)
for labels in ax1.get_xticklabels():
    labels.set_ha('right')

df.groupby('Rating')['JP_Sales'].sum().sort_values(ascending=False).head(4).plot(kind='bar', ax = ax2)
ax2.yaxis.set_major_formatter(plt.FuncFormatter(lambda y, pos: f'${int(y)}M'))
ax2.set_title('Videogame Sales in Japan by Rating')
ax2.set_xticklabels(ax2.get_xticklabels(), rotation=45)
fig.suptitle('1980-2009', fontsize = 18)
plt.tight_layout()
plt.show()
```

```python
fig2, (ax1, ax2) = plt.subplots(1, 2, figsize = (10, 6))
df.groupby('Genre')['EU_Sales'].sum().sort_values(ascending= False).plot(kind= 'bar', ax = ax1)
ax1.yaxis.set_major_formatter(plt.FuncFormatter(lambda y, pos: f'${int(y)}M'))
ax1.set_title('Videogame Sales in Europe by Genre')
ax1.set_xticklabels(ax1.get_xticklabels(), rotation=45)
for labels in ax1.get_xticklabels():
    labels.set_ha('right')

df.groupby('Rating')['EU_Sales'].sum().sort_values(ascending= False).head(4).plot(kind= 'bar', ax = ax2)
ax2.yaxis.set_major_formatter(plt.FuncFormatter(lambda y, pos: f'${int(y)}M'))
ax2.set_title('Videogame Sales in Europe by Rating')
ax2.set_xticklabels(ax2.get_xticklabels(), rotation=45)
fig2.suptitle('1980-2009', fontsize = 18)
plt.tight_layout()
plt.show()
```
